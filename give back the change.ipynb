```python
cost=int(input("enter the cost of the product"))
amount_given=int(input("enter the amount to be given:"))
result=amount_given-cost


def change():
    if (cost<amount_given):
        print ("thank you,visit again")
        A=0
        count=0
        if (result%100==0):        
            a=int(result/100)
            if(a!=0):
                print('No.of 100 NOTES =')
                print(a)
        elif(result%100>0):
            A=result%100
            Aa=int((result-A)/100)
            if(Aa!=0):
                print('No.of 100 NOTES =')
                print(Aa)
            count+=1
        if(count>0):
            if ((A%50>=0)):
                B=int(A%50)
                b=int(A/50)
                if(b!=0):
                    print('No.of 50 NOTES =')
                    print(b)
                if(A%50>0):
                    count+=1
            elif((result%50>=0)):              
                B=(result%50)                
                b=int((result-B)/50)
                if(b!=0):
                    print('No.of 50 NOTES =')
                    print(b)
                if(result%50>0):
                    count+=1
        if(count>1):
            if ((B%20>=0)):
                C=int(B%20)
                c=int(B/20)
                if(c!=0):
                    print('No.of 20 NOTES =')
                    print(c)
                if(B%20>0):
                    count+=1
            elif((result%20>=0)):              
                C=(result%20)                
                c=int((result-C)/20)
                if(c!=0):
                    print('No.of 20 NOTES =')
                    print(c)
                if(result%20>0):
                    count+=1
        if(count>2):
            if ((C%10>=0)):
                D=int(C%10)
                d=int(C/10)
                if(d!=0):
                    print('No.of 10 NOTES =')
                    print(d)
                if(C%10>0):
                    count+=1
            elif((result%10>=0)):              
                D=(result%10)                
                d=int((result-D)/10)
                if(d!=0):
                    print('No.of 10 NOTES =')
                    print(d)
                if(result%10>0):
                    count+=1
        if(count>3):
            if ((D%5>=0)):
                E=int(D%5)
                e=int(D/5)
                if(e!=0):
                    print('No.of 5 NOTES =')
                    print(e)
                if(D%5>0):
                    count+=1
            elif((result%5>=0)):              
                E=(result%5)                
                e=int((result-E)/5)
                if(e!=0):
                    print('No.of 5 NOTES =')
                    print(e)
                if(result%5>0):
                    count+=1
        if(count>4):
            if ((E%2>=0)):
                F=int(E%2)
                f=int(E/2)
                if(f!=0):
                    print('No.of 2 COINS =')
                    print(f)
                if(E%2>0):
                    count+=1
            elif((result%2>=0)):              
                F=(result%2)                
                f=int((result-F)/2)
                if(f!=0):
                    print('No.of 2 COINS =')
                    print(f)
                if(result%2>0):
                    count+=1
        if(count>5):
            print('no. of 1 COIN =')
            print("1")

        print("CHANGE =",result)
    else:
        print('please tender more money!')
        exit()        
change()
```
